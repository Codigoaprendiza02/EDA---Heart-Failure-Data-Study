**Understanding distributions of various features.**

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 12))

ax1.hist(df['age'], color='green')
ax1.set(xlabel = 'Age',
        ylabel = 'Frequency',
        title = 'Age Distribution');

ax2.hist(df['creatinine_phosphokinase'], color='blue')
ax2.set(xlabel = 'creatinine_phosphokinase',
       ylabel = 'Frequency',
       title = 'Creatinine_phosphokinase distribution');

ax3.hist(df['ejection_fraction'], color='red')
ax3.set(xlabel = 'ejection_fraction',
        ylabel = 'Frequency',
        title = 'Ejection_fraction distribution');

ax4.hist(df['platelets'], color='black')
ax4.set(xlabel = 'platelets',
        ylabel = 'Frequency',
        title = 'Platelets distribution');
```
