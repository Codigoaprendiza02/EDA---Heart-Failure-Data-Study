# Heart Failure Clinical Dataset – Basic EDA

## Project Overview
This project performs a **very basic Exploratory Data Analysis (EDA)** on the **Heart Failure Clinical Records Dataset** from Kaggle.  
The aim is to understand the dataset structure, explore simple patterns, and identify factors related to heart failure outcomes.

---

## Dataset Details
- **Source:** Kaggle – Heart Failure Clinical Records Dataset  
- **Records:** 299 patients  
- **Features:** 13 clinical features + 1 target variable  

### Target Variable
- **DEATH_EVENT**
  - `1` → Patient died during follow-up
  - `0` → Patient survived
---

##  Feature Summary
- `age` – Age of patient
- `anaemia` – Anaemia status (0/1)
- `creatinine_phosphokinase` – CPK enzyme level
- `diabetes` – Diabetes status (0/1)
- `ejection_fraction` – Percentage of blood leaving heart
- `high_blood_pressure` – Hypertension status (0/1)
- `platelets` – Platelet count
- `serum_creatinine` – Serum creatinine level
- `serum_sodium` – Serum sodium level
- `sex` – Gender (0 = female, 1 = male)
- `smoking` – Smoking status (0/1)
- `time` – Follow-up time (days)

**Understanding distributions of various features.**

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 12))

ax1.hist(df['age'], color='green')
ax1.set(xlabel = 'Age',
        ylabel = 'Frequency',
        title = 'Age Distribution');

ax2.hist(df['creatinine_phosphokinase'], color='blue')
ax2.set(xlabel = 'creatinine_phosphokinase',
       ylabel = 'Frequency',
       title = 'Creatinine_phosphokinase distribution');

ax3.hist(df['ejection_fraction'], color='red')
ax3.set(xlabel = 'ejection_fraction',
        ylabel = 'Frequency',
        title = 'Ejection_fraction distribution');

ax4.hist(df['platelets'], color='black')
ax4.set(xlabel = 'platelets',
        ylabel = 'Frequency',
        title = 'Platelets distribution');
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(5, 5))

sns.histplot(data=df, x=df['ejection_fraction'], hue=df['DEATH_EVENT'], multiple='dodge', bins=20)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 8))
sns.scatterplot(data=df, x="age", y="creatinine_phosphokinase", hue="DEATH_EVENT", ax=ax1);
ax1.axhline(df["creatinine_phosphokinase"].mean(), linestyle="--", color="black")
sns.scatterplot(data=df, x="age", y="platelets", hue="DEATH_EVENT", palette=["green", "red"], ax=ax2);
ax2.axhline(df["platelets"].mean(), linestyle="--", color="black")
sns.scatterplot(data=df, x="age", y="ejection_fraction", hue="DEATH_EVENT", palette=["black", "orange"], ax=ax3);
ax3.axhline(df["ejection_fraction"].mean(), linestyle="--", color="red")
sns.scatterplot(data=df, x="age", y="serum_sodium", palette=["purple", "pink"], hue="DEATH_EVENT", ax=ax4);
ax4.axhline(df["serum_sodium"].mean(), linestyle='--', color="black")
```

```python
plt.figure(figsize=(10, 10))
dataframe1 = pd.DataFrame({
    'high_blood_pressure' : df["high_blood_pressure"],
    'anaemia' : df["anaemia"],
    "diabetes" : df["diabetes"],
    'sex' : df["sex"],
    'age' : df['age'],
})
dataframe1.head()
correlation_mat1 = dataframe1.corr()
```

```python
sns.heatmap(correlation_mat1, annot=True, linewidths=.5, cmap="plasma")
```

```python
plt.figure(figsize=(12, 12))
dataframe2 = pd.DataFrame({
    'age': df['age'],
    'platelets' : df['platelets'],
    'ejection_fraction' : df['ejection_fraction'],
    'death event' : df['DEATH_EVENT'],
})
dataframe2.head()   
```

```python
correlation_mat2 = dataframe2.corr()
sns.heatmap(correlation_mat2, annot=True, linewidth=.5, cmap='viridis')
```

```python
plt.figure(figsize=(12, 12))
dataframe3 = pd.DataFrame({
    'creatinine_phosphokinase' : df['creatinine_phosphokinase'],
    'serum_sodium' : df['serum_sodium'],
    'serum_creatinine' : df['serum_creatinine'],
    'death_event' : df['DEATH_EVENT']
})
dataframe3.head()
```

```python
correlation_matrix3 = dataframe3.corr()
sns.heatmap(correlation_matrix3, annot=True, linewidth=.5, cmap='Greens')
```
