# Heart Failure Clinical Dataset – Basic EDA

## Project Overview
This project performs a **very basic Exploratory Data Analysis (EDA)** on the **Heart Failure Clinical Records Dataset** from Kaggle.  
The aim is to understand the dataset structure, explore simple patterns, and identify factors related to heart failure outcomes.

---

## Dataset Details
- **Source:** Kaggle – Heart Failure Clinical Records Dataset  
- **Records:** 299 patients  
- **Features:** 13 clinical features + 1 target variable  

### Target Variable
- **DEATH_EVENT**
  - `1` → Patient died during follow-up
  - `0` → Patient survived
---

##  Feature Summary
- `age` – Age of patient
- `anaemia` – Anaemia status (0/1)
- `creatinine_phosphokinase` – CPK enzyme level
- `diabetes` – Diabetes status (0/1)
- `ejection_fraction` – Percentage of blood leaving heart
- `high_blood_pressure` – Hypertension status (0/1)
- `platelets` – Platelet count
- `serum_creatinine` – Serum creatinine level
- `serum_sodium` – Serum sodium level
- `sex` – Gender (0 = female, 1 = male)
- `smoking` – Smoking status (0/1)
- `time` – Follow-up time (days)

# Removal of outliers

1. I will use **trimming method** to remove the outliers where **less number of outliers**.
2. I will use **capping method** to remove the outlier where **huge number of outliers**.

### 1. Creatinine_phosphokinase

```python
cp_percentile25 = df['creatinine_phosphokinase'].quantile(0.25)
cp_percentile75 = df['creatinine_phosphokinase'].quantile(0.75)
cp_iqr = cp_percentile75 - cp_percentile25
cp_upperlimit = cp_percentile75 + 1.5*cp_iqr
cp_lowerlimit = cp_percentile25 - 1.5*cp_iqr
df[(df['creatinine_phosphokinase']>cp_upperlimit) | (df['creatinine_phosphokinase']<cp_lowerlimit)]
```

```python
df['creatinine_phosphokinase'] = df['creatinine_phosphokinase'].clip(
    lower=cp_lowerlimit,
    upper=cp_upperlimit
)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10, 5))
sns.boxplot(df['creatinine_phosphokinase'], color = 'blue', ax=ax1)
sns.histplot(df['creatinine_phosphokinase'], color = 'blue', ax=ax2, kde=True)
```

### 2. Serum_creatinine

```python
sc_percentile25 = df['serum_creatinine'].quantile(0.25)
sc_percentile75 = df['serum_creatinine'].quantile(0.75)
sc_iqr = sc_percentile75 - sc_percentile25
sc_upperlimit = sc_percentile75 + 1.5*sc_iqr
sc_lowerlimit = sc_percentile25 - 1.5*sc_iqr
df[(df['serum_creatinine']>sc_upperlimit) | (df['serum_creatinine']<sc_lowerlimit)]
```

```python
df['serum_creatinine'] = df['serum_creatinine'].clip(
    lower=sc_lowerlimit,
    upper=sc_upperlimit
)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(df['serum_creatinine'], color='green', ax=ax1)
sns.histplot(df['serum_creatinine'], color='green', ax=ax2, kde=True)
```

### 3. Platelets 

```python
pl_percentile25 = df['platelets'].quantile(0.25)
pl_percentile75 = df['platelets'].quantile(0.75)
pl_iqr = pl_percentile75 - pl_percentile25
pl_upperlimit = pl_percentile25 + 1.5*pl_iqr
pl_lowerlimit = pl_percentile75 - 1.5*pl_iqr
df[(df['platelets']>pl_upperlimit) | (df['platelets']<pl_lowerlimit)]
```

```python
df['platelets'] = df['platelets'].clip(
    lower=pl_lowerlimit,
    upper=pl_upperlimit
)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(df['platelets'], color='red', ax=ax1)
sns.histplot(df['platelets'], color='red', ax=ax2, kde=True)
```

### 4. Serum_sodium

```python
ss_percentile25 = df['serum_sodium'].quantile(0.25)
ss_percentile75 = df['serum_sodium'].quantile(0.75)
ss_iqr = ss_percentile75 - ss_percentile25
ss_upperlimit = ss_percentile75 + 1.5*ss_iqr
ss_lowerlimit = ss_percentile25 - 1.5*ss_iqr
df = df[(df['serum_sodium']<ss_upperlimit) & (df['serum_sodium']>ss_lowerlimit)]
df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(df['serum_sodium'], color='orange', ax=ax1)
sns.histplot(df['serum_sodium'], color='orange', ax=ax2, kde=True)
```

### 5. Ejection_fraction

```python
ef_percentile25 = df['ejection_fraction'].quantile(0.25)
ef_percentile75 = df['ejection_fraction'].quantile(0.75)
ef_iqr = ef_percentile75 - ef_percentile25
ef_upperlimit = ef_percentile75 + 1.5*ef_iqr
ef_lowerlimit = ef_percentile25 - 1.5*ef_iqr
df = df[(df['ejection_fraction']<ef_upperlimit) & (df['ejection_fraction']>ef_lowerlimit)]
df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(df['ejection_fraction'], color='black', ax=ax1)
sns.histplot(df['ejection_fraction'], 
             color='black',
             ax=ax2, 
             kde=True, 
)
```

# Correlation matrix

Let's see the features are correlated with each other and "DEATH_EVENT" using **correlation matrix** and **heatmaps**.

```python
plt.figure(figsize=(15, 15))
dataframe1 = pd.DataFrame({
    'high_blood_pressure' : df["high_blood_pressure"],
    'anaemia' : df["anaemia"],
    "diabetes" : df["diabetes"],
    'sex' : df["sex"],
    'age' : df['age'],
    'death_event' : df['DEATH_EVENT'],
})
dataframe1.head()
correlation_mat1 = dataframe1.corr()
```

```python
sns.heatmap(correlation_mat1, annot=True, linewidths=.5, cmap="plasma")
```

**Insights from above heatmap**

1. **Age** and **Death event** shows the highest +ve relation with each other, i.e. 0.25. 
2. **Diabetes** and **Sex** shows the highest -ve relation with each other, i.e. -0.095.
3. **High_blood_pressure** and **Sex** show the good +ve relation with each other, i.e. 0.093. 
4. **Diabetes** show negative correlation with all the features. 
5. **Age** is showing correlation with every feature in the above heatmap. 
6. **Diabetes** is having very less relation with **death_event**
